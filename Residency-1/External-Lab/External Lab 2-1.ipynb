### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
# Assumes that 1 is Male and 2 is Female
all_males = HR[HR['Gender'] == 1]['Gender']
print("Total number of males in the company = ", all_males.count())
```

```python
# Assumes that 1 is Male and 2 is Female
all_females = HR[HR['Gender'] == 2]['Gender']
print("Total number of females in the company = ", all_females.count())
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
quantile_25_75 = HR.Age.quantile([.25, .75])
print("25th and 75th Quartile data is below")
print(quantile_25_75)
print('\n')
print('IQR for Age is below', quantile_25_75[0.75] - quantile_25_75[0.25])
```

### Find out the mean, median and variance for the column - DailyRate 

```python
print("Mean, Median and Variance for DailyRate columns are indicated below:")
print("Mean = ", HR.DailyRate.mean())
print("Median = ", HR.DailyRate.median())
print("Variance = ", HR.DailyRate.var())
```
